```python
import csv
from math import sqrt
import pandas as pd
places=[]
lat=[]
lon=[]
a=[]
file = open('LatLon.csv')
reader = csv.reader(file)
next(reader,None)
for row in reader:
    places.append(str(row[0]))
    lat.append(float(row[1])) 
    lon.append(float(row[2]))  
for i in places:
    for j in places:
        ii=places.index(i)
        jj=places.index(j)
        x=sqrt(((lat[ii]-lat[jj])**2)+((lon[ii]-lon[jj])**2))
        a.append([i,j,x])
#print(a)
df = pd.DataFrame(a, columns=['x','y','z'])

# create pivot table with x and rows, y and columns, z as values
h=df.pivot_table(values='z', index='x', columns='y')
print(h['Bangalore']['Agra'])
```
